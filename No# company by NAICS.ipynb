```python
from pandas import Series, DataFrame
import pandas as pd 
import numpy as np 
```

```python
d16 = pd.read_csv("/Users/kimcw/Desktop/23/33100806.csv") 
```

```python
d = d16[
    ((d16["North American Industry Classification System (NAICS)"] == "Utilities [22]") 
    | (d16["North American Industry Classification System (NAICS)"] == "Manufacturing [31-33]") 
    | (d16["North American Industry Classification System (NAICS)"] == "Transportation and warehousing [48-49]") 
    | (d16["North American Industry Classification System (NAICS)"] == "Retail trade [44-45]") 
    | (d16["North American Industry Classification System (NAICS)"] == "Professional, scientific and technical services [54]") 
    | (d16["North American Industry Classification System (NAICS)"] == "Public administration[91]"))
    & ((d16["GEO"] == "Canada") | (d16["GEO"] == "Ontario") | (d16["GEO"] == "Quebec") | (d16["GEO"] == "British Columbia"))
]

d.head(10)
```

```python
sort_order = ["Utilities [22]", "Manufacturing [31-33]",
              "Transportation and warehousing [48-49]",
              "Retail trade [44-45]",
              "Professional, scientific and technical services [54]","Public administration[91]"]
d['sort_order'] = d['North American Industry Classification System (NAICS)'].apply(lambda x: sort_order.index(x))

d = d.drop(columns=["DGUID", "UOM", "UOM_ID", "SCALAR_FACTOR", "SCALAR_ID", "VECTOR"])
d = d.drop(columns=['STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS'], errors='ignore')

d = d[d['Employment size'] == "Total, with employees"]
```
