```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np 
```

```python
sch = pd.read_excel("/Users/chaewon/Desktop/exdata/exschool.xlsx")
sch.head()
```

```python
sch["TOEFL iBT Total"] = pd.to_numeric(sch["TOEFL iBT Total"], errors='coerce')
sch["TOEFL Reading"] = pd.to_numeric(sch["TOEFL Reading"], errors='coerce')
sch["TOEFL Listening"] = pd.to_numeric(sch["TOEFL Listening"], errors='coerce')
sch["TOEFL Speaking"] = pd.to_numeric(sch["TOEFL Speaking"], errors='coerce')
sch["TOEFL Writing"] = pd.to_numeric(sch["TOEFL Writing"], errors='coerce')
sch["Placement Number (Exchange Quota)"] =  pd.to_numeric(sch["Placement Number (Exchange Quota)"], errors='coerce')
sch["TOEFL iBT Total"].fillna(0, inplace=True)
sch["TOEFL Reading"].fillna(0, inplace=True)
sch["TOEFL Listening"].fillna(0, inplace=True)
sch["TOEFL Speaking"].fillna(0, inplace=True)
sch["TOEFL Writing"].fillna(0, inplace=True)
```

```python
sch1 = sch[(sch["Country"]=="United States")
    & (sch["TOEFL iBT Total"] <= 89)
    & (sch["TOEFL Reading"] <= 25)
    & (sch["TOEFL Listening"] <= 26)
    & (sch["TOEFL Speaking"] <= 18)
    & (sch["TOEFL Writing"] <= 20)
    & (sch["Placement Number (Exchange Quota)"] >= 5)]

sch1
```

```python
sch1.to_csv("/Users/chaewon/Desktop/exdata/2025 America 교환교 리스트 정리.csv", encoding='utf-8-sig', index=False)
```
